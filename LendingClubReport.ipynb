# P2P平台 LendingClub 贷款数据分析  

## 概要
- **概念区分**
  - 小贷公司：即小额贷款公司，是指由自然人、企业法人与其他社会组织投资设立，不吸收公众存款，经营小额贷款业务的有限责任公司或股份有限公司。
  - 网络小贷：小额贷款公司主要通过网络平台获取借款客户，综合运用网络平台积累的客户经营、网络消费、网络交易等行为数据、即时场景信息等，分析评定借款客户信用风险，确定授信方式和额度，并在线上完成贷款申请、风险审核、贷款审批、贷款发放和贷款回收等全流程的小额贷款业务。
  - 网贷平台：即网络借贷信息中介机构，是指依法设立，专门从事网络借贷信息中介业务活动的金融信息中介公司。该类机构以互联网为主要渠道，为借款人与出借人实现直接借贷提供信息搜集、信息公布、资信评估、信息交互、借贷撮合等服务。
  - 区别：网络小贷是小贷公司的一种，是放贷机构，二者均由地方金融办负责牌照发放和监管事宜，只是在经营区域上有限制；网贷平台定位为信息中介机构，不具有放贷资质，但网贷平台的借款产品和网络小贷的产品均属于网络借贷的范畴。  
  
  
- **项目背景和目的**
  - LendingClub是美国规模最大的P2P网贷平台，以互联网金融创新模式建立了一个比传统借贷机构更高效的在线交易平台，投资和贷款流程高度自动化，利用在线数据和技术快速评估风险，确定贷款信用评级和利率，检测欺诈风险。通过LendingClub平台，投资者可选择多元化投资组合，获得具竞争力的回报，贷款人则可获得较低的贷款利率，降低借贷成本。 
  - 本项目通过分析LC官网个人贷款业务数据，探索LendingClub个人贷款业务特点、贷款客户特征、贷款违约影响因素，了解网贷平台在商业模式、业务结构、风控管理等方面与传统小贷公司的区别，并且通过构建贷款违约预测模型，了解互联网创新技术、线上风控模式在互联网小额贷款业务领域的实际应用。  
  
  
- **总结**
  - LendingClub的营业收入以贷款交易费为主，较为依赖贷款交易规模，平台个人贷款业务为客户提供小额借贷服务，业务规模整体呈增长趋势，受季节性因素影响，有一定的波动性。
  - 平台内部风险控制机制较为成熟，对贷款申请人的债务收入比、征信记录等进行准入审核，贷款客户整体收入状况、资金状况、信用记录良好。
  - 违约贷款的客户收入主要集中于低收入水平，近期存在频繁征信查询记录和违约事项的违约风险较大。
  - 平台实行差异化风险定价策略，贷款利率的设定综合考虑信用风险因素，贷款信用评级等级越低，利率越高。
  - 平台信用评级模型较为成熟，贷款信用评级等级越低，贷款违约率越高。  
     
     
- **建议**
  - 从风险识别的角度来看：LendingClub线上风控模式较为依赖信用评分或年收入等量化指标，对于风险评估，还应考虑美国消费者提前消费、过度消费的观念和季节性因素的影响。收入越高的个人消费支出越大，可能更容易陷入资不抵债的困境，个人申请贷款主要用于债务合并、信用卡还款，贷款客户财务风险较大。在第四季度，消费者对假期消费带来的财务压力，导致贷款逾期、坏账的可能性增大。
  - 从信用评级标准的角度来看：根据2017年财务报告显示，LendingClub2015-2017年贷款交易规模和营业收入增长缓慢，平台通过调整信用评级标准和定价策略以维持交易规模发展，贷款逾期率自2015年起超过美国银行消费贷款平均水平。LendingClub保持业务增长态势的同时也应落实风控机制，保证贷款质量，控制违约风险，关注经营效益。
  - 从高风险客群的角度来看：一般情况下，P2P贷款客户主要为难以达到传统银行信贷标准的高风险人群，信贷资金的回款取决于客户的还款能力和还款意愿。LendingClub应根据不同贷款信用评级的客户特征信息，对高违约风险的细分客群加强甄别与监察，及时预警贷款逾期的风险，积极开展催收工作，降低坏账风险。

### 1.1 项目背景：
- LendingClub是美国规模最大的P2P网贷平台，创立于2006年，总部位于旧金山，2014年12月在纽约证券交易所上市，成为当年最大的科技股IPO。LendingClub以互联网金融创新模式建立了一个比传统借贷机构更高效的在线交易平台，投资和贷款流程高度自动化，利用在线数据和技术快速评估风险，确定贷款信用评级和利率，检测欺诈风险。通过LendingClub平台，投资者可选择多元化投资组合，获得具竞争力的回报，贷款人则可获得较低的贷款利率，降低借贷成本。  
- 目前，LendingClub提供的贷款服务包括个人贷款、小企业商业贷款、汽车贷款和抵押、教育和医疗信贷服务。根据2017年财务报告显示，LendingClub2017年放贷总规模达到89.87亿美元，其中个人贷款服务放贷总额为81.31亿美元，占比90%，是推动平台业务规模增长的关键服务。根据2018年第二季度业绩报告显示，2018年上半年平台放贷总规模达到51.24亿美元，同比增长25%，环比增长5%。

- 综合以上分析，贷款违约影响因素有以下特点：
  - 贷款金额、利率越高，期限越长，贷款违约的风险越大；
  - 贷款信用评级等级越低，贷款违约的可能性越大；
  - 客户的收入、资金、信用状况会影响违约风险，违约贷款的客户收入主要集中于低收入水平，近期存在频繁征信查询记录和违约事项的违约风险较大；
  - 客户的工作年限对违约率的影响差别不大，贷款申请信息不完整的客户违约风险较大。

## 四、构建贷款违约预测模型

### 4.1 算法选择

- 本文对贷款违约影响因素进行探索分析，构建模型预测贷款是否会违约，已知预测的目标变量为贷款状态——是否会违约，属于二元分类问题，选择监督学习中的分类算法——逻辑回归。逻辑回归是应用非常广泛的一个分类机器学习算法，它将数据拟合到一个logistic函数中，从而能够完成对事件发生的概率进行预测。

### 4.6 模型评估

- 混淆矩阵：混淆矩阵是机器学习中对监督学习的分类算法进行评估的常用工具，将真实值与预测值对比，利用评估指标反映模型分类的效果，矩阵的行表示真实类别，矩阵的列表示预测类别。
  - 对于二分类问题，混淆矩阵的四个矩阵元素分别为：
    - True Positive(TP)：真正例，真实类别为正例（正常），预测类别为正例（正常）；
    - False Positive(FP)：假正例，真实类别为负例（违约），预测类别为正例（正常）；
    - False Negative(FN)：假负例，真实类别为正例（正常），预测类别为负例（违约）；
    - True Negative(TN)：真负例，真实类别为负例（违约），预测类别为负例（违约）。
  - 混淆矩阵评估指标：
    - 准确率：Accuracy(ACC) = (TP+TN)/(TP+TN+FP+FN)，预测正确的样本数占总样本数的比例；
    - 精确率：Precision(PPV,Positive predictive value) = TP/(TP+FP)，预测为正例的样本中，预测正确的样本所占比例；
    - 召回率：Recall(TPR,True positive rate) = TP/(TP+FN)，真实为正例的样本中，正确预测为正例的样本所占比例；
    - 假正率：Fall-out(FPR,False positive rate) = FP/(FP+TN)，真实为负例的样本中，错误预测为正例的样本所占比例。
